```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Process RSSI data in .csv file
# Returns object with statistical data
def process_data(horientation):
    data_directory = "data/" + horientation + "/1"
    xn = 6 # Number of x axis limit
    yn = 3 # Number of y axis limit

    data = np.zeros((yn,xn))

    for x in range(xn):
        for y in range(yn):
            try:
                f = open(data_directory + '/' + str(x) + str(y) + '.csv', 'r')
            except:
                print("Could not read file " + str(x) + str(y))

            content = f.read().split('\n')
            rssis = np.array(content[0:-1]).astype(float)
            data[y,x] = rssis.max()
            
    return data

data_obj = {
    "horizontal": process_data("horizontal"),
    "vertical": process_data("vertical"),
    "lateral": process_data("lateral")
}
```

```python
def plot_heat(ax, d, horientation):
    ax.imshow(d, origin="lower", vmin=-80, vmax=-20)
    ax.set_xticks(np.arange(xn))
    ax.set_yticks(np.arange(yn))
    ax.set_title(horientation)
    ax.axis('on')
    
    for i in range(d.shape[0]):
        for j in range(d.shape[1]):
            text = ax.text(j, i, d[i, j],
                           ha="center", va="center", color="w")

# Plot heatmap
fig, ax = plt.subplots(1, 3, figsize=(20,30), sharex='all')

plot_heat(ax[0], data_obj["horizontal"], "Horizontal")
plot_heat(ax[1], data_obj["vertical"], "Vertical")
plot_heat(ax[2], data_obj["lateral"], "Lateral")

plt.savefig("RSSI_shelve1.png", bbox_inches='tight')
```
